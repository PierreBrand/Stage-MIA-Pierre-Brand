### $V$ - MFA - données simulées

#### 5.1 - MFA for PPCA

```python
#Simulation des données

thetas = fc.sim_param_mix_1(K1,D1,L1,s3=0.5)
Z_1, omega_1, Y_1 = fc.sim_data_mix_1(thetas,N1)
```

```python
#Estimation de thetas et Z dans le cas où omega est connu

thetas_hat, Z_hat, omega_hat = fc.MFA_1(Y_1,L1,K_omega=omega_1)
Y_hat = fc.MFA_rec_1(thetas_hat,Z_hat,omega_1)
```

```python
#Représentation graphique dans le cas où omega est connu

tri_Y = fc.tri(Y_1,omega_1)
tri_Y_hat = fc.tri(Y_hat,omega_1)
colors = ['#802020','#E08080','#208020','#80E080','#202080','#8080E0','#808020','#E0E080','#802080','#E080E0','#208080','#80E0E0','#202020','#E0E0E0']

for j in range(int(D1/2)):
    
    plt.figure()
    
    for k in range(K1):
        plt.scatter(tri_Y[k][:,2*j],tri_Y[k][:,2*j+1],label='$Y$',color=colors[2*k])
        plt.scatter(tri_Y_hat[k][:,2*j],tri_Y_hat[k][:,2*j+1],label='$\hat{Y}$',color=colors[2*k+1])

    for i in range(N1):
        plt.plot([Y_1[i][2*j],Y_hat[i][2*j]],[Y_1[i][2*j+1],Y_hat[i][2*j+1]],color='black')

    plt.legend()
    plt.show()
```

```python
#Estimation de thetas et Z dans le cas où omega est inconnu

thetas_hat, Z_hat, omega_hat = fc.MFA_1(Y_1,L1,K_omega=K1)
Y_hat = fc.MFA_rec_1(thetas_hat,Z_hat,omega_hat)
```

```python
#Représentation graphique dans le cas où omega est inconnu

s_opt = fc.perm_opt(omega_1,omega_hat)
s_omega = np.array([s_opt[o] for o in omega_1])
tri_Y = fc.tri(Y_1,s_omega)
tri_Y_hat = fc.tri(Y_hat,omega_hat)
colors = ['#802020','#E08080','#208020','#80E080','#202080','#8080E0','#808020','#E0E080','#802080','#E080E0','#208080','#80E0E0','#202020','#E0E0E0']

for j in range(int(D1/2)):
    
    plt.figure()
    
    for k in range(K1):
        plt.scatter(tri_Y[k][:,2*j],tri_Y[k][:,2*j+1],label='$Y$',color=colors[2*k])
        plt.scatter(tri_Y_hat[k][:,2*j],tri_Y_hat[k][:,2*j+1],label='$\hat{Y}$',color=colors[2*k+1])

    for i in range(N1):
        plt.plot([Y_1[i][2*j],Y_hat[i][2*j]],[Y_1[i][2*j+1],Y_hat[i][2*j+1]],color='black')

    plt.legend()
    plt.show()
```
